```python
# ============================================
# Pacientes atendidos por servicio
# ============================================

import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# 1️⃣ Rutas correctas
ruta_datos = Path.cwd().parent / "database" / "pacientes_clean.csv"
ruta_salida = Path.cwd().parent / "reports"  # sin la barra inicial

# 2️⃣ Cargar datos limpios
df = pd.read_csv(ruta_datos)
print(f"✅ Archivo cargado: {ruta_datos}")

# 3️⃣ Crear carpeta de salida (si no existe)
ruta_salida.mkdir(parents=True, exist_ok=True)

# 4️⃣ Gráfico de barras: pacientes por servicio
count_serv = df['servicio'].value_counts()
count_serv.plot.bar()

plt.title('Pacientes atendidos por servicio')
plt.xlabel('Servicio')
plt.ylabel('N° de pacientes')
plt.tight_layout()

# 5️⃣ Guardar imagen en carpeta reports (rutas bien concatenadas)
ruta_imagen = ruta_salida / "pacientes_atendidos_por_servicio.png"
plt.savefig(ruta_imagen)
plt.show()

print(f"📊 Gráfico guardado correctamente en: {ruta_imagen}")
```

```python
# 
# Generamos los graficos de distribucion por sexo y edades, y el reporte de promedio de montos por distrito
# ============================================


import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd

# Rutas
ruta_datos = Path.cwd().parent / "database" / "pacientes_clean.csv"
ruta_salida = Path.cwd().parent / "reports"

# Cargar datos
df = pd.read_csv(ruta_datos)

# Asegurar que la carpeta reports existe
ruta_salida.mkdir(parents=True, exist_ok=True)

# --------------------------
# 1️⃣ Gráfico circular: distribución por sexo
# --------------------------
plt.figure()  # crear figura nueva
df['sexo'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribución por sexo')
plt.ylabel('')  # opcional: quitar etiqueta lateral
ruta_imagen = ruta_salida / "distribucion_por_sexo.png"
plt.savefig(ruta_imagen)
plt.show()
print(f"📊 Gráfico guardado en: {ruta_imagen}")

# --------------------------
# 2️⃣ Histograma de edades
# --------------------------
plt.figure()
df['edad'].hist(bins=15)
plt.title('Distribución de edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.tight_layout()
ruta_imagen = ruta_salida / "distribucion_de_edades.png"
plt.savefig(ruta_imagen)
plt.show()
print(f"📊 Gráfico guardado en: {ruta_imagen}")

# --------------------------
# 3️⃣ Tabla/Reporte: promedio de montos por distrito
# --------------------------
prom = (
    df.groupby('distrito')['monto']
    .mean()
    .reset_index()
    .sort_values('monto', ascending=False)
)

ruta_csv = ruta_salida / "promedio_monto_por_distrito.csv"
prom.to_csv(ruta_csv, index=False)
print(f"📋 Reporte guardado en: {ruta_csv}")

# Mostrar los primeros resultados en el notebook
prom.head()

```
